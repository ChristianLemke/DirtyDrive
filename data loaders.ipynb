```python
#dates
import datetime

days = []
for x in range(1,10):
    days.append(datetime.datetime(2016,12,x,0,0,0))
```

```python
# threaded data downloader
from DirtyDrive import data_loader
from DirtyDrive import db_saver
import datetime
import threading

def download_by_threads(filenames):
    # subparts
    n= 8
    l = filenames
    for sublist in [l[i:i + n] for i in xrange(0, len(l), n)]:
        print(sublist)
        # thread download 
        threads = [threading.Thread(target=data_loader.download_file, args=(filename,)) 
                   for filename in sublist]

        for thread in threads:
            thread.start()
        for thread in threads:
            thread.join()

db_saver = db_saver.db_saver()
data_loader = data_loader.data_loader()

for day_x in days:
    db_saver.load_filenames()
    
    #minutes = [0,15,30,45]
    minutes = [0,10,20,30,40,50]
    
    db_saver.filter_dates(day_x, minutes)
    
    files = [x for x in db_saver.filtered_page_dates if not data_loader.file_exists(x)]
    download_by_threads(files)

```
